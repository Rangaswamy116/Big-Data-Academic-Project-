# Hypothesis testing

```python
from scipy.stats import ttest_ind

# Perform a t-test to compare two groups
group1 = df[df['carrier'] == 'Group1']['arr_delay']
group2 = df[df['carrier'] == 'Group2']['arr_delay']
t_stat, p_value = ttest_ind(group1, group2)
print('T-Statistic:', t_stat)
print('P-Value:', p_value)
```

The p-value represents the likelihood of detecting the observed t-statistic if the null hypothesis (no difference between the group means) were true. The t-statistic quantifies the difference between the means of the two groups.

You may determine the statistical significance of the difference between the two groups' arrival delays by looking at the t-statistic and p-value.

```python

from pyspark.sql.functions import avg, desc, row_number
from pyspark.sql.window import Window

delays = df_Dealy.groupBy("year", "airport",).agg(avg("arr_del15").alias("arr_flights"))
window_spec = Window.partitionBy("year").orderBy(desc("arr_flights"))
ranked_delays = delays.withColumn("rank", row_number().over(window_spec))

# Filter top 10 origins for each year
top_origins = ranked_delays.filter("rank <= 10")

# Show the top 10 origin airports with highest average delay year-wise
top_origins.orderBy("year", "rank").show()
```

In the DataFrame 'df_Dealy', the given code determines the average arrival delay ('arr_del15') for each combination of year and airport. Then, using a window function, it ranks the airports within each year according to the typical delay.
The output lists the top 10 origin airports with the longest average delays by year, rank, airport, and average delay. You may use this data to determine which airports had the most delays each year, which will be useful for assessing and tracking delays at certain sources over time.


```python
import matplotlib.pyplot as plt
top_origins_pd = top_origins.toPandas()
from pyspark.sql.functions import count
import seaborn as sns

years = top_origins_pd["year"].unique()
for year in years:
    year_data = top_origins_pd[top_origins_pd["year"] == year]
    origins = year_data["airport"]
    flight_counts = year_data["arr_flights"]

    plt.figure(figsize=(12, 6))
    sns.lineplot(x=origins, y=flight_counts, marker="o")
    plt.xlabel("ORIGIN")
    plt.ylabel("FLIGHT COUNT")
    plt.title(f"Top 15 Origins with Highest Flight Counts in {year}")
    plt.xticks(rotation=90)
    plt.show()
```

The modified code sample uses line plots to display the flight numbers for the top 15 origin airports with the greatest annual average delays. It uses a top_origins DataFrame that has been transformed into a top_origins_pd Pandas DataFrame and integrates the Seaborn library for visualization.
For each year, the output creates line graphs that show the number of flights for the top 15 origin airports with the longest average delays. These line graphs let you compare the number of flights from various origin airports across time and spot any patterns or fluctuations in the number of flights.

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.feature import Imputer, StandardScaler
```

```python
# Split the data into training and testing sets
train_ratio = 0.8
test_ratio = 0.2
train_data, test_data = df_Dealy.randomSplit([train_ratio, test_ratio], seed=42)
```

```python
# Encoding categorical variables
indexers = [StringIndexer(inputCol=col, outputCol=col+"_index").fit(train_data) for col in ["carrier", "airport"]]
encoders = [OneHotEncoder(inputCol=col+"_index", outputCol=col+"_encoded") for col in ["carrier", "airport"]]
pipeline = Pipeline(stages=indexers + encoders)
train_data = pipeline.fit(train_data).transform(train_data)
test_data = pipeline.fit(test_data).transform(test_data)
# Print the output
train_data.show()
test_data.show()
```

In the output, the training and testing data have been converted, and the categorical variables "carrier" and "airport" have been one-hot encoded and indexes. The indexed and encoded columns have "_index" and "_encoded" prefixes, respectively, and are attached to the original data. The categorical variables may now be represented as numerical features that can be included into machine learning models thanks to the encoding procedure.

```python
from pyspark.sql.functions import desc, col

# Group the data by airport and count the total number of flights
flight_counts = df_Dealy.groupBy("airport_name").agg(count("*").alias("total_count"))

# Add the counts for the destination airports
flight_counts = flight_counts.union(
    df_Dealy.groupBy("arr_flights").agg(count("*").alias("total_count"))
)

# Sum the counts for each airport and select the busiest one
df_airport= flight_counts.groupBy("airport_name").agg({"total_count": "sum"}) \
    .orderBy(desc("sum(total_count)")).first()["airport_name"]
print("Austin:", df_airport)
```

```python
from pyspark.sql.functions import avg, desc

# Filter out cancelled flights
airline_df = df_Dealy.filter(df_Dealy["arr_cancelled"] == 0)

# Group by origin and calculate average departure delay
avg_delayed= airline_df.groupBy("airport") \
                                   .agg(avg("arr_del15").alias("arr_delay")) \
                                   .orderBy(desc("arr_delay"))

# Filter for only airports with average delay greater than 0 minutes
df_delayed = avg_delayed.filter("arr_delay > 0") \
                                           .limit(10)

df_delayed.show()
```

The given code computes the typical departure delay for each airport and filters out cancelled flights from the DataFrame df_Dealy. The top 10 airports with the highest average delay are then chosen after filtering out airports having an average delay higher than 0 minutes.

```python
from pyspark.ml.feature import StringIndexer

cols_for_encoding = ['year', 'month', 'carrier', 'airport', 'airport_name', 'arr_flights', 'arr_del15', 'carrier_ct',
                     'weather_ct', 'nas_ct', 'security_ct', 'late_aircraft_ct', 'arr_cancelled', 'arr_diverted',
                     'arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']

# Iterate through the columns and apply StringIndexer to encode the categorical data
indexers = []
for column in cols_for_encoding:
    indexer = StringIndexer(inputCol=column, outputCol=column + '_index')
    df_Dealy = indexer.fit(df_Dealy).transform(df_Dealy)
    indexers.append(indexer)

# Show the encoded DataFrame
df_Dealy.show()
```

In the given code, categorical data in the DataFrame df_Dealy is encoded using StringIndexer from the pyspark.ml.feature module. It performs the StringIndexer transformation on each of the columns listed in the cols_for_encoding list as it iterates over them. The "_index" suffix is applied to the modified columns before they are included in the DataFrame. Finally, the show() method is used to display the encoded DataFrame.
The DataFrame df_Dealy with its original columns and matching encoded columns, which have the "_index" suffix, are displayed in the output.
